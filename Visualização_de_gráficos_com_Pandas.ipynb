# Criando um gráfico simples


```
import matplotlib.pyplot as plt

eixo_x = [1,2,5]
eixo_y = [2,3,7]

plt.plot(eixo_x, eixo_y)
plt.show()
```

# Legendas e títulos

```
import matplotlib.pyplot as plt

eixo_x = [1,2,5]
eixo_y = [2,3,7]

plt.plot(eixo_x, eixo_y)

#Titulo
plt.title("Meu primeiro gráfico Pyton")

#Legendas
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.show()
```

# Gráfico de barras

```
import matplotlib.pyplot as plt

eixo_x = [1,2,3]
eixo_y = [2,3,7]

plt.bar(eixo_x, eixo_y)

#Titulo
plt.title("Meu primeiro gráfico Pyton")

#Legendas
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.show()
```

# Comparação com gráficos de barras

```
import matplotlib.pyplot as plt

eixo_x_1 = [4,2,3]
eixo_y_1 = [3,3,7]

eixo_x_2 = [1,5,3]
eixo_y_2 = [1,8,7]

#Plotando o gráfico e adiconando a legenda para cada um
plt.bar(eixo_x_1, eixo_y_1, label= "Grupo 1")
plt.bar(eixo_x_2, eixo_y_2, label= "Grupo 2")

#Titulo
plt.title("Meu primeiro gráfico Pyton")

#Legendas
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

#Plotar alegenda
plt.legend()

plt.show()
```

# Gráfico Scatterplot ou gráfico de dispersão

```
import matplotlib.pyplot as plt

eixo_x = [1,3,5,7,9]
eixo_y = [2,3,7,1,0] 

#Plotando o gráfico Scatterplot ligando os pontos plotando o gráfico de linhas
plt.scatter(eixo_x, eixo_y, label= "Grupo 1", color="red")
plt.plot(eixo_x, eixo_y)

#Titulo
plt.title("Gráfico de Scatterplot")

#Legendas
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

#Plotar a legenda
plt.legend()

plt.show()
```

# Marcadores

```
import matplotlib.pyplot as plt

eixo_x = [1,3,5,7,9]
eixo_y = [2,3,7,1,0] 

#Plotando o gráfico Scatterplot ligando os pontos plotando o gráfico de linhas
plt.scatter(eixo_x, eixo_y, label= "Grupo 1", color="g", marker=".", s=100)
plt.plot(eixo_x, eixo_y, linestyle=":")

#Titulo
plt.title("Gráfico de Scatterplot")

#Legendas
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

#Plotar a legenda
plt.legend()

plt.show()
```

# Salvar figura do gráfico gerado

```
import matplotlib.pyplot as plt

eixo_x = [1,3,5,7,9]
eixo_y = [2,3,7,1,0] 

#Plotando o gráfico Scatterplot ligando os pontos plotando o gráfico de linhas
plt.scatter(eixo_x, eixo_y, label= "Grupo 1", color="g", marker=".", s=100)
plt.plot(eixo_x, eixo_y, linestyle=":")

#Titulo
plt.title("Gráfico de Scatterplot")

#Legendas
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

#Plotar a legenda
plt.legend()

plt.show()
#PDF
plt.savefig("figura1.pdf")

#Formato png com resolução maior
plt.savefig("figura1.png", dpi=300)
```
